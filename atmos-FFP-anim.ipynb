```python
import os
from datetime import datetime
import pandas as pd
import numpy as np

import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.animation as animation

plt.style.use('dark_background')
```

```python
imgfiles = [f for f in os.listdir('./output/') if 'ATMOS_FFP_2024-04-09_' in f]
imgfiles.sort()
imgfiles
```

```python
fig, ax = plt.subplots()

frames = []

for i in range(len(imgfiles)):
    img = mpimg.imread(os.path.join('./output', imgfiles[i]))
        
    frame = plt.imshow(img, animated=True)
    if i==0:
        plt.imshow(img)
    frames.append([frame])
    
ani = animation.ArtistAnimation(fig, frames, interval=400, blit=True, repeat_delay=5000)

plt.axis('off')
plt.tight_layout()

ani.save('./output/ATMOS_FFP_anim_2024-04-09.gif')
plt.show()
```
